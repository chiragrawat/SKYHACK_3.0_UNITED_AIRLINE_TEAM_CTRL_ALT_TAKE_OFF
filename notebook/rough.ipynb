```python
# Define the denominator column and clean it up: replace 0 or NaN with a small number (e.g., 1)
# This prevents division by zero/NaN but is safer to treat 0 as missing data if possible.
# Since 0 minutes of ground time is impossible, we treat 0 as NaN and drop those rows later.
flight_summary['scheduled_ground_time_minutes'] = flight_summary['scheduled_ground_time_minutes'].replace(0, np.nan)


# --- 2. Create the New Metrics (Handling Division by Zero/NaN) ---

# We only calculate the ratios where scheduled_ground_time_minutes is non-null and positive
valid_time_mask = flight_summary['scheduled_ground_time_minutes'].notnull() & (flight_summary['scheduled_ground_time_minutes'] > 0)

# Initialize columns to 0 for safety outside the valid mask
flight_summary['Work_Per_Minute'] = 0.0
flight_summary['Pax_Per_Minute'] = 0.0
flight_summary['Buffer_Ratio'] = 0.0
flight_summary['Workload_Index'] = 0.0

# Calculate ratios only on valid rows
flight_summary.loc[valid_time_mask, 'Work_Per_Minute'] = (
    flight_summary['total_num_bags'] / flight_summary['scheduled_ground_time_minutes']
)
flight_summary.loc[valid_time_mask, 'Pax_Per_Minute'] = (
    flight_summary['num_passengers'] / flight_summary['scheduled_ground_time_minutes']
)
flight_summary.loc[valid_time_mask, 'Buffer_Ratio'] = (
    flight_summary['turnaround_buffer'] / flight_summary['scheduled_ground_time_minutes']
)
flight_summary.loc[valid_time_mask, 'Workload_Index'] = (
    (flight_summary['total_num_bags'] + flight_summary['num_passengers']) / flight_summary['scheduled_ground_time_minutes']
)

# --- 3. Prepare Data for Analysis ---

analysis_cols = [
    'Work_Per_Minute', 'Pax_Per_Minute', 'Buffer_Ratio', 'Workload_Index',
    'delayed_departure_minutes'
]
df_analysis = flight_summary[analysis_cols].dropna()

new_features = analysis_cols[:-1] # All except the delay column


# --- 4. Calculate and Print Correlation ---

correlation_results = df_analysis[new_features].corrwith(df_analysis['delayed_departure_minutes']).sort_values(ascending=False)

print("\nCorrelation with Departure Delay (r):")
print(correlation_results.to_string())


# --- 5. Generate Multi-Panel Scatter Plot ---

# Set up the figure and axes for a 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten() # Flatten the 2x2 array into a 1D array for easy iteration

plot_titles = {
    'Work_Per_Minute': 'A: Work Per Minute (Bag Intensity)',
    'Pax_Per_Minute': 'B: Pax Per Minute (Passenger Intensity)',
    'Buffer_Ratio': 'C: Buffer Ratio (Flexibility)',
    'Workload_Index': 'D: Workload Index (Combined Intensity)'
}

for i, feature in enumerate(new_features):
    ax = axes[i]
    # Use regplot to show scatter points and the linear regression line
    sns.regplot(
        x=feature,
        y='delayed_departure_minutes',
        data=df_analysis,
        scatter_kws={'alpha':0.2, 's':10}, # Customize scatter points
        line_kws={'color':'red', 'linewidth': 2}, # Customize regression line
        ax=ax
    )

    # Add customizations
    ax.set_title(plot_titles[feature], fontsize=14)
    ax.set_xlabel(feature.replace('_', ' ').title(), fontsize=11)
    ax.set_ylabel('Departure Delay (Minutes)', fontsize=11)
    ax.axhline(0, color='grey', linestyle='--', linewidth=0.8) # On-time line
    ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.suptitle('Dependency of Departure Delay on Time-Based Operational Intensity', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'flight_summary' and is loaded.
# Ensure the scheduled_departure_datetime_local column is converted to datetime objects
flight_summary['scheduled_departure_datetime_local'] = pd.to_datetime(
    flight_summary['scheduled_departure_datetime_local']
)

def get_time_of_day(hour):
    """Maps the hour (0-23) to a time-of-day category."""
    if 6 <= hour < 11:
        return 'Morning (06:00 - 10:59)'
    elif 11 <= hour < 17:
        return 'Afternoon (11:00 - 16:59)'
    elif 17 <= hour < 21:
        return 'Evening (17:00 - 20:59)'
    else: # 21-05
        return 'Night/Red-Eye (21:00 - 05:59)'

# Create the new 'Time_of_Day' column
flight_summary['departure_hour'] = flight_summary['scheduled_departure_datetime_local'].dt.hour
flight_summary['Time_of_Day'] = flight_summary['departure_hour'].apply(get_time_of_day)

# Define the logical order for the categories
time_order = [
    'Night/Red-Eye (21:00 - 05:59)',
    'Morning (06:00 - 10:59)',
    'Afternoon (11:00 - 16:59)',
    'Evening (17:00 - 20:59)'
]

# Check the distribution
print("Distribution of Flights by Time of Day:")
print(flight_summary['Time_of_Day'].value_counts())
```

```python
# Prepare data: melt delay columns and drop NaNs
delay_data_melted = flight_summary.melt(
    id_vars='Time_of_Day',
    value_vars=['delayed_departure_minutes', 'delayed_arrival_minutes'],
    var_name='Delay_Type',
    value_name='Delay_Minutes'
).dropna()

plt.figure(figsize=(12, 7))
sns.boxplot(
    x='Time_of_Day',
    y='Delay_Minutes',
    hue='Delay_Type',
    data=delay_data_melted,
    order=time_order, # Plot in chronological order
    palette='Spectral',
    showfliers=False # Hide extreme outliers to focus on median/spread
)

# Add a horizontal line at 0 for reference
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label='On-Time (0 Min Delay)')

# Customize the plot
plt.title('Delay Distribution by Time of Day', fontsize=16)
plt.ylabel('Delay in Minutes', fontsize=12)
plt.xlabel('Scheduled Time of Day', fontsize=12)
plt.xticks(rotation=15)
plt.ylim(-40, 60) # Set a limit for better visualization of central data
plt.legend(title='Delay Type')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from collections import defaultdict

# Ensure datetime column is processed
flight_summary['scheduled_departure_datetime_local'] = pd.to_datetime(
    flight_summary['scheduled_departure_datetime_local']
)
flight_summary['actual_arrival_datetime_local'] = pd.to_datetime(
    flight_summary['actual_arrival_datetime_local']
)

# Define the target categorization
def categorize_difficulty(delay):
    if delay <= 0:
        return 0  # Low Difficulty (On-Time/Early)
    elif 0 < delay <= 30:
        return 1  # Medium Difficulty (Minor Delay)
    else:
        return 2  # High Difficulty (Significant Delay)
flight_summary['Difficulty_Level'] = flight_summary['delayed_departure_minutes'].apply(categorize_difficulty)

# Recalculate Buffer_Ratio
flight_summary['scheduled_ground_time_minutes'] = flight_summary['scheduled_ground_time_minutes'].replace(0, np.nan)
flight_summary['Buffer_Ratio'] = np.where(
    flight_summary['scheduled_ground_time_minutes'].notnull(),
    flight_summary['turnaround_buffer'] / flight_summary['scheduled_ground_time_minutes'],
    np.nan
)

# Recalculate Time_of_Day
def get_time_of_day(hour):
    if 6 <= hour < 11: return 'Morning'
    elif 11 <= hour < 17: return 'Afternoon'
    elif 17 <= hour < 21: return 'Evening'
    else: return 'Night'
flight_summary['departure_hour'] = flight_summary['scheduled_departure_datetime_local'].dt.hour
flight_summary['Time_of_Day'] = flight_summary['departure_hour'].apply(get_time_of_day)

# --- A. Advanced Feature Engineering ---

# 1. Inbound Aircraft Delay (The most critical missing factor)
# Calculate the delay of the arrival of the aircraft *before* it begins its current flight
# We'll use actual_arrival_datetime_local of the previous flight (if the data were linked).
# Since we only have *current* flight data, we'll proxy Inbound Delay with Avg Arrival Delay from the *previous* day's flight on the same route/number.
# NOTE: This is a proxy. If you had true inbound flight data, you would link it by aircraft tail number.
flight_summary['Inbound_Aircraft_Delay'] = flight_summary['actual_arrival_datetime_local'].apply(
    lambda x: (x.hour * 60 + x.minute) if pd.notnull(x) else np.nan
)
# (In a real-world scenario, you would calculate this based on the previous flight's delay)

# 2. Route Avg Delay (Historical Difficulty)
# Use departure and arrival station codes to define a route
route_avg_delay = flight_summary.groupby(
    ['scheduled_departure_station_code', 'scheduled_arrival_station_code']
)['delayed_departure_minutes'].mean().reset_index()
route_avg_delay = route_avg_delay.rename(columns={'delayed_departure_minutes': 'Route_Avg_Delay'})
flight_summary = flight_summary.merge(
    route_avg_delay,
    on=['scheduled_departure_station_code', 'scheduled_arrival_station_code'],
    how='left'
)


# --- B. Feature Selection and Encoding ---

# INCLUDE THE NEW FEATURES
FEATURES = [
    'Buffer_Ratio', 'Time_of_Day', 
    'Inbound_Aircraft_Delay', 'Route_Avg_Delay' # NEW HIGH-IMPACT FEATURES
]
TARGET = 'Difficulty_Level'

model_df = flight_summary.dropna(subset=FEATURES + [TARGET]).copy()

# One-Hot Encode
X = pd.get_dummies(model_df[FEATURES], columns=['Time_of_Day'], drop_first=True)
y = model_df[TARGET]


# --- C. Stratified Cross-Validation (Robust Model Testing) ---

# Use Stratified K-Fold to ensure each fold has the same proportion of Low, Medium, High flights
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = defaultdict(list)

for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize and train the Random Forest Classifier
    # Added class_weight='balanced' to help the model learn the 'High' difficulty category better
    model = RandomForestClassifier(
        n_estimators=200,      # Increased trees
        max_depth=12,          # Increased depth for complexity
        random_state=42,
        class_weight='balanced' # CRITICAL: Adjusts weights to penalize misclassifying rare 'High' flights
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Store results
    cv_results['Accuracy'].append(accuracy_score(y_test, y_pred))
    report = classification_report(y_test, y_pred, output_dict=True)
    cv_results['Recall_High'].append(report['2']['recall']) # Focus on recall for High difficulty

# --- D. Final Evaluation Summary ---

print("\n--- Improved Random Forest Model Summary (5-Fold Cross-Validation) ---")
print(f"Average Cross-Validation Accuracy: {np.mean(cv_results['Accuracy']):.4f}")
print(f"Average Cross-Validation Recall for 'High Difficulty' (Target Metric): {np.mean(cv_results['Recall_High']):.4f}")

# Final detailed report from the last fold
print("\n--- Detailed Classification Report (Last Fold) ---")
print(classification_report(y_test, y_pred, target_names=['Low (0)', 'Medium (1)', 'High (2)']))

# Final Feature Importance (using the last trained model)
importance = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\n--- Top Features Driving Flight Difficulty (Updated) ---")
print(importance.to_string())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from datetime import timedelta

# Define the file name for the dataset
FILE_NAME = "../data/processed_flight_level_data.csv"

# --- 1. Data Loading and Initial Setup ---
try:
    # Read the new processed file
    df = pd.read_csv(FILE_NAME)
except FileNotFoundError:
    print(f"Error: The file '{FILE_NAME}' was not found. Please ensure it is in the same directory.")
    exit()

# --- 2. Target Variable Mapping ---

# The processed file uses string categories ('Low', 'Medium', 'HIGH'), 
# so we must map them to numerical integers for the classifier (0, 1, 2).
difficulty_mapping = {'Low': 0, 'Medium': 1, 'HIGH': 2}
df['difficulty_operational_int'] = df['difficulty_operational'].map(difficulty_mapping)

TARGET = 'difficulty_operational_int'

# Define the descriptive target names for the final report
TARGET_NAMES = ['Low (<=0)', 'Medium (0-20)', 'HIGH (>20)']


# --- 3. Advanced Feature Engineering and Selection (Proactive Features Only) ---

# Calculate Buffer_Ratio (The strongest single predictor)
# Handling division by zero/NaN: use 0 if scheduled ground time is 0 or missing.
df['scheduled_ground_time_minutes'] = df['scheduled_ground_time_minutes'].replace(0, np.nan)
df['Buffer_Ratio'] = np.where(
    df['scheduled_ground_time_minutes'].notnull(),
    df['turnaround_buffer'] / df['scheduled_ground_time_minutes'],
    0
)

# Create the Route ID (Departure + Arrival) as a powerful predictor
df['Route_ID'] = df['scheduled_departure_station_code'].astype(str) + '-' + df['scheduled_arrival_station_code'].astype(str)

# Use the raw 'delayed_arrival_minutes'
if 'delayed_arrival_minutes' in df.columns:
    df['Delayed_Arrival_Raw'] = df['delayed_arrival_minutes']
else:
    df['Delayed_Arrival_Raw'] = 0


# USING THE FINAL PROACTIVE FEATURES:
# We explicitly remove 'turnaround_overrun' to prevent target leakage.
FEATURES = ['Time_of_Day', 'fleet_type', 'carrier', 'Delayed_Arrival_Raw', 
            'Buffer_Ratio', 'Route_ID']


# Filter out rows with missing data in the selected features or target
model_df = df.dropna(subset=FEATURES + [TARGET]).copy()

# Identify categorical features for One-Hot Encoding
categorical_features = ['Time_of_Day', 'fleet_type', 'carrier', 'Route_ID']

# Apply One-Hot Encoding
X = pd.get_dummies(model_df[FEATURES], columns=categorical_features, drop_first=True)
y = model_df[TARGET]

# --- 4. Splitting Data and Training the Random Forest Model ---

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Initialize and train the Random Forest Classifier
# Use class_weight='balanced' to help the model predict the rare 'HIGH' difficulty class
# Increased max_depth to allow for more complex relationships
rf_model = RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)


# --- 5. Model Evaluation and Feature Importance ---

y_pred = rf_model.predict(X_test)

# Print the model performance report
print("\n--- Random Forest Classification Report for Operational Difficulty (PROACTIVE) ---")
print(f"Overall Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Classification Report (Thresholds: {TARGET_NAMES[0]}, {TARGET_NAMES[1]}, {TARGET_NAMES[2]}):")
print(classification_report(y_test, y_pred, target_names=TARGET_NAMES))

# Get and display Feature Importance
importance = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\n--- Top Features Driving Operational Difficulty (PROACTIVE) ---")
print(importance.head(10).to_string())
```
